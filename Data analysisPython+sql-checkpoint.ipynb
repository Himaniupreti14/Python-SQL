```python
#extract file from zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip') 
zip_ref.extractall() # extract file to dir
zip_ref.close() # close file
```

```python
# read data from the file and handle the null vales
import pandas as pd
df= pd.read_csv('orders.csv.zip', na_values = ['Not Available', 'unknown'])
df.head(20)
df['Ship Mode'].unique()
```

```python
#rename column names and maske them lower case and replace the spaces with the undersore

df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ', '_')
df.columns
```

```python
## derive new columns discount,sales price and profit

df['discount'] = df['list_price']*df['discount_percent']*.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
df
```

```python
#convert order date from object data type to date time 
df['order_date']=pd.to_datetime(df['order_date'],format = "%Y-%m-%d")
df.dtypes

```

```python
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
        
```
